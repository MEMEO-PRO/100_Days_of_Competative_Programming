```python
# Let's load the dataset and perform basic EDA (Exploratory Data Analysis)
# The user requested line graph, scatter plot, histogram, qq plot, and descriptive statistics.

import pandas as pd

# Load the dataset
file_path = '/content/kwc.csv'
df = pd.read_csv(file_path)

# Display basic info and first few rows
df_info = df.info()
df_head = df.head()

df_info, df_head
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats

# Convert TxnDate to datetime
df['TxnDate'] = pd.to_datetime(df['TxnDate'], format='%d-%b-%y')

# Descriptive statistics
descriptive_stats = df.describe()

# Set up plots
plt.figure(figsize=(14, 10))

# Line plot of Consumption over time
plt.subplot(2, 2, 1)
plt.plot(df['TxnDate'], df['Consumption'], color='blue')
plt.title('Line Plot of Consumption Over Time')
plt.xlabel('Date')
plt.ylabel('Consumption')

# Scatter plot
plt.subplot(2, 2, 2)
sns.scatterplot(x=df['TxnDate'], y=df['Consumption'], color='orange')
plt.title('Scatter Plot of Consumption')
plt.xlabel('Date')
plt.ylabel('Consumption')

# Histogram of Consumption
plt.subplot(2, 2, 3)
sns.histplot(df['Consumption'], bins=30, kde=True, color='green')
plt.title('Histogram of Consumption')
plt.xlabel('Consumption')

# QQ Plot
plt.subplot(2, 2, 4)
stats.probplot(df['Consumption'], dist="norm", plot=plt)
plt.title('QQ Plot of Consumption')

plt.tight_layout()
plt.show()

descriptive_stats
```

```python
# Stationarity check using rolling statistics and ADF test
from statsmodels.tsa.stattools import adfuller

# Calculate rolling statistics (moving average and standard deviation)
rolling_mean = df['Consumption'].rolling(window=30).mean()
rolling_std = df['Consumption'].rolling(window=30).std()

# Perform ADF test
adf_result = adfuller(df['Consumption'])

# Plot rolling mean and standard deviation
plt.figure(figsize=(10, 6))
plt.plot(df['TxnDate'], df['Consumption'], color='blue', label='Original Consumption')
plt.plot(df['TxnDate'], rolling_mean, color='red', label='Rolling Mean')
plt.plot(df['TxnDate'], rolling_std, color='black', label='Rolling Std Dev')
plt.title('Rolling Mean & Standard Deviation')
plt.xlabel('Date')
plt.ylabel('Consumption')
plt.legend(loc='best')
plt.show()

# ADF test result
adf_result
```

```python
# prompt: i want to perform eda on bmw.csv dataset plot line graph, histogram having data in form Date	Open	High	Low	Close	Adj Close	Volume
# 2016-08-24	77.250000	78.029999	76.690002	77.510002	61.236576	971047 and scatter plot histogram, qq plot and descriptive statistics

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats

# Load the dataset
file_path = '/content/bmw.csv'  # Replace with your actual file path
df = pd.read_csv(file_path)

# Convert Date to datetime
df['Date'] = pd.to_datetime(df['Date'])

# Descriptive statistics
descriptive_stats = df.describe()

# Set up plots
plt.figure(figsize=(14, 10))

# Line plot of Close over time
plt.subplot(2, 2, 1)
plt.plot(df['Date'], df['Close'], color='blue')
plt.title('Line Plot of Closing Price Over Time')
plt.xlabel('Date')
plt.ylabel('Closing Price')

# Scatter plot
plt.subplot(2, 2, 2)
sns.scatterplot(x=df['Date'], y=df['Close'], color='orange')
plt.title('Scatter Plot of Closing Price')
plt.xlabel('Date')
plt.ylabel('Closing Price')

# Histogram of Close
plt.subplot(2, 2, 3)
sns.histplot(df['Close'], bins=30, kde=True, color='green')
plt.title('Histogram of Closing Price')
plt.xlabel('Closing Price')

# QQ Plot
plt.subplot(2, 2, 4)
stats.probplot(df['Close'], dist="norm", plot=plt)
plt.title('QQ Plot of Closing Price')

plt.tight_layout()
plt.show()

descriptive_stats
```

```python
# prompt: perform stationarity and ADF test on bmw.csv dataset

# Stationarity check using rolling statistics and ADF test

# Calculate rolling statistics (moving average and standard deviation)
rolling_mean = df['Close'].rolling(window=30).mean()
rolling_std = df['Close'].rolling(window=30).std()

# Perform ADF test
adf_result = adfuller(df['Close'])

# Plot rolling mean and standard deviation
plt.figure(figsize=(10, 6))
plt.plot(df['Date'], df['Close'], color='blue', label='Original Close Price')
plt.plot(df['Date'], rolling_mean, color='red', label='Rolling Mean')
plt.plot(df['Date'], rolling_std, color='black', label='Rolling Std Dev')
plt.title('Rolling Mean & Standard Deviation of Close Price')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.legend(loc='best')
plt.show()

# ADF test result
print("ADF Test Results:")
print(f"ADF Statistic: {adf_result[0]}")
print(f"p-value: {adf_result[1]}")
print(f"Critical Values:")
for key, value in adf_result[4].items():
    print(f"\t{key}: {value}")

```

```python
# prompt: replace missing values and remove stationarity in the data of bmw.csv

# Load the dataset
file_path = '/content/bmw.csv'  # Replace with your actual file path
df = pd.read_csv(file_path)

# Convert Date to datetime
df['Date'] = pd.to_datetime(df['Date'])

# Check for missing values
missing_values = df.isnull().sum()
print("Missing Values:\n", missing_values)

# Replace missing values (if any) with the mean of the column
df.fillna(df.mean(), inplace=True)

# Remove stationarity (if present) using differencing
df['Close_diff'] = df['Close'].diff()

# Drop the first row as it will have a NaN value after differencing
df.dropna(inplace=True)

# Check for stationarity again using ADF test
adf_result_diff = adfuller(df['Close_diff'])

print("\nADF Test Results after differencing:")
print(f"ADF Statistic: {adf_result_diff[0]}")
print(f"p-value: {adf_result_diff[1]}")
print(f"Critical Values:")
for key, value in adf_result_diff[4].items():
    print(f"\t{key}: {value}")

# Plot the differenced data
plt.figure(figsize=(10, 6))
plt.plot(df['Date'], df['Close_diff'], color='blue', label='Differenced Close Price')
plt.title('Differenced Closing Price Over Time')
plt.xlabel('Date')
plt.ylabel('Differenced Closing Price')
plt.legend(loc='best')
plt.show()
```

```python
# prompt: check for seasonality and seasonality decomposition from
# df['Date'], df['Close_diff'],

from statsmodels.tsa.seasonal import seasonal_decompose

# Assuming 'df' and 'df['Date']', 'df['Close_diff']' are defined

# Perform seasonal decomposition
result = seasonal_decompose(df['Close_diff'], model='additive', period=30)  # Assuming a monthly seasonality, adjust period as needed

# Plot the decomposition components
plt.figure(figsize=(12, 8))

plt.subplot(411)
plt.plot(df['Date'], df['Close_diff'], label='Original')
plt.legend(loc='best')
plt.title('Seasonal Decomposition of Differenced Closing Price')

plt.subplot(412)
plt.plot(result.trend, label='Trend')
plt.legend(loc='best')

plt.subplot(413)
plt.plot(result.seasonal, label='Seasonality')
plt.legend(loc='best')

plt.subplot(414)
plt.plot(result.resid, label='Residual')
plt.legend(loc='best')

plt.tight_layout()
plt.show()

```

```python
# prompt: perform ACF and PACF plots of above

from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

# Assuming 'df['Close_diff']' is your differenced time series data

# ACF Plot
plt.figure(figsize=(12, 6))
plot_acf(df['Close_diff'], lags=30)  # Adjust lags as needed
plt.title('Autocorrelation Function (ACF)')
plt.show()

# PACF Plot
plt.figure(figsize=(12, 6))
plot_pacf(df['Close_diff'], lags=30)  # Adjust lags as needed
plt.title('Partial Autocorrelation Function (PACF)')
plt.show()
```

```python
# prompt: apply AR(1) on the above dataset NotImplementedError: AR has been removed from statsmodels and replaced with statsmodels.tsa.ar_model.AutoReg.

from statsmodels.tsa.ar_model import AutoReg

# Assuming df['Close_diff'] is your stationary time series data

# Fit the AR(1) model
model = AutoReg(df['Close_diff'], lags=1)  # Specify lags=1 for AR(1)
results = model.fit()

# Print model summary
print(results.summary())

# Forecast future values
forecast_steps = 10  # Number of steps to forecast
forecast = results.predict(start=len(df), end=len(df) + forecast_steps - 1)

# Plot the forecast
plt.figure(figsize=(10, 6))
plt.plot(df['Date'], df['Close_diff'], label='Original Data')
plt.plot(pd.date_range(start=df['Date'].iloc[-1] + pd.DateOffset(days=1), periods=forecast_steps), forecast, label='AR(1) Forecast')
plt.title('AR(1) Model Forecast')
plt.xlabel('Date')
plt.ylabel('Differenced Close Price')
plt.legend()
plt.show()
```
